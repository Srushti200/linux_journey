```python
import numpy as np
import matplotlib.pyplot as plt

# Define input data and labels
X = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
Y = np.array([-1, -1, -1, 1])

# Initialize weights and bias
w = np.zeros(X.shape[1])
b = 0

# Perceptron training loop
for _ in range(6):
    for i in range(X.shape[0]):
        y_pred = np.sign(np.dot(X[i], w) + b)
        
        if y_pred != Y[i]:
            w += 0.3 * Y[i] * X[i]
            b += 0.3 * Y[i]

# Create mesh grid for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max,0.01),
                     np.arange(y_min, y_max,0.01))

# Predict class for each grid point
Z = np.sign(np.dot(np.c_[xx.ravel(), yy.ravel()], w) + b)
Z = Z.reshape(xx.shape)

# Plot decision boundary with custom colors
plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')  # Change colormap to 'coolwarm'

# Plot data points with different colors
scatter = plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', edgecolors='k', s=100)  
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Decision Regions')



# Show the plot
plt.show()
```
